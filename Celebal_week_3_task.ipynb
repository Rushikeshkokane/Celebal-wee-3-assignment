Iterating (for loop) - On List, Tuple, Set, Dictionary

```python
#List
n_list = ["Abhay","Bhavesh","Siddhant","Kiran"]
for x in n_list:
  print(x)

print("")

#tuple
a_tuple=[39,"Rushi","mango"]
for y in a_tuple:
  print(y)

print("")


#dictionary
c_dict = {
    'a1':'Rahul',
    'b2':'Shubham',
    'c3':'Kunal',
    'd4':'Rushi',
}

for x,y in c_dict.items():
  print(x,y)

print("")
#Set
x_set = {24,'Rushi','Kunal'}
for z in x_set:
  print(z)
```

Indexing

```python
#indexing in list
fruits=['apple','banana','pineapple','cherry']
x=fruits.index('pineapple')
print(x)

print("")
#indexing in tuple
tuple=("c","e","l","e","b","a","l")
print(tuple.index("c"))
print(tuple.index("l"))
```

Membership tests(in ,not in)

```python
#first we will see use of in membership operator

name=['R','U','S','H','I']
ch = input('Please Enter a Capital Letter:')

if ch in name:
    print('You entered a correct character')
else:
    print('You entered a wrong character')
```

```python
# Membership test(not in)
list = [10, 20, 30, 40, 50 ];
x = input('Please Enter the number:')

if ( x not in list ):
   print("x is NOT present in given list")
else:
   print("x is  present in given list")
```

Accessing Nested List using Slicing

```python
#List
l1 = [21,43,15,[34,64,77]]
print(l1)
print(l1[3][0])

print("")

#Tuple
s1 = (12,43,['rushi','kunsl'])

print(s1[2][0])
```
